```python
my_ts = ts6['item 10960'].to_frame()
my_ts.plot()
```

Нужно стабилизировать дисперсию

```python
my_ts.dropna().values.min()
```

Чтобы бокс-кокс сработал, поднимем ряд

```python
my_ts = my_ts.dropna() + 1.
```

```python
my_ts['box-cox'], lmbda = boxcox(my_ts)
```

```python
my_ts['box-cox'].to_frame().plot()
```

Теперь дисперсия намного стабильнее, но ряд нельзя считать стационарным на уровне значимости 0.01.

```python
thr = int(0.8 * len(my_ts))
train, test = my_ts.head(thr), my_ts.tail(len(ts6) - thr)
```

```python
sm.tsa.adfuller(train['box-cox'].ravel())
```

На уровне значимости $0.05$ наш ряд можно принять стационарным. Посмотрим на автокорелляции:

```python
sm.graphics.tsa.plot_acf(train.values.ravel(), lags=50)
```

```python
sm.graphics.tsa.plot_pacf(train.values.ravel(), lags=50)
```

Выглядит не очень приятно. Продифференцируем.

```python
diff1 = ((train['box-cox'] - train['box-cox'].shift(1)).dropna()).to_frame()
diff1.plot()
```

```python
sm.tsa.stattools.adfuller(diff1.values.ravel())
```

Дики-Фуллер говорит, что теперь все ок.

Посмотрим на acf+pacf, чтобы найти хорошие значения $p$ и $q$ ($d$ полагаем равным $1$)

```python
sm.graphics.tsa.plot_acf(diff1.values.ravel(), lags=50)
```

```python
sm.graphics.tsa.plot_pacf(diff1.values.ravel(), lags=50)
```

Учитывая последние статзначимые значения, стоит рассмотреть $p\sim4, q\sim3$

```python
pdq = product(range(3,6), [1], range(2,5))
best_params = (0, 0, 0)
best_score = np.inf
for param in pdq:
    try:
        model = ARIMA(diff1['box-cox'].values, param).fit(disp=-1)
        if(model.aic < best_score):
            best_params = param
            best_score = model.aic
    except:
        continue
print(best_params, best_score)
```

Итого, лучшая модель - $ARIMA(4,1,2)$. Проверим ее остатки.

```python
best_model = ARIMA(diff1['box-cox'].values, (best_params)).fit()
```

```python
residuals = best_model.resid
plt.plot(best_model.resid)
```

```python
sm.tsa.adfuller(residuals)
```

```python
ttest_1samp(residuals, 0.)
```

```python
shapiro(residuals)
```

```python
probplot(residuals, plot=plt)
plt.show()
```

```python
sm.graphics.tsa.plot_acf(residuals, lags=50)
```

```python
sm.graphics.tsa.plot_pacf(residuals, lags=50)
```

Итого, остатки нормальны, стационарны, неавтокореллированы и с нулевым средним. Что мы и хотели.

Посмотрим на прогноз нашей модели:

```python
print(test.index[0], test.index[-1])
```

```python
forecast = best_model.forecast(steps=10)[0]
```

```python
test['forecast'] = forecast
```
